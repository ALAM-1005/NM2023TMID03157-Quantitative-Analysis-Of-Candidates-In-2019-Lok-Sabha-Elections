# Data Preparation

```python
# Finding Corelation between datasets 
df1 = df.select_dtypes(["number"])
df1.corr()
```

```python
# Renaming Column
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace('\n','_')
df
```

```python
# Finding null values
df.isnull().sum()
```

```python
df1 = df[df['party']!='NOTA']
df1.isnull().sum()
```

```python
# Replacing Null and Unwanted Values with 0
df['criminal_cases']= df['criminal_cases'].replace({'Not Available':0})
df = df.fillna(0)
df['criminal_cases'] = pd.to_numeric(df['criminal_cases'],errors='ignore').astype(np.int64)
df['age']=df['age'].apply(lambda x:round(x))
df['age'] = pd.to_numeric(df['age'])
```

# Understanding Relation among Dataset

```python
# Understanding relation using HeatMap
df1 = df.select_dtypes("number")
sns.heatmap(df1.corr(),annot=True, cmap= "YlGnBu")
```

```python
# Paiplot for getting Individual Relation
sns.pairplot(df.select_dtypes(['number']), hue = 'age')
```

```python
# Barplot: state vs Criminal cases
sns.barplot(x='state' ,y ='criminal_cases', data =df)
plt.xticks(rotation = 90)
```

```python
# Statewise no. of constituencies
st=df.groupby('state').apply(lambda x:x['constituency'].nunique()).reset_index(name='constituency')
st
```

```python
sns.barplot(x= 'state', y = 'constituency', data = st)
plt.xticks(rotation = 90) 
plt.show()
```

```python
# Relation between No. of Candidates and their Education
df.drop(df[df['education']=='Post Graduate\n'].index, inplace=True)
sns.countplot(x='education', data = df)
plt.xticks(rotation = 90)
```

```python
df1=df[df['party']!='NOTA']
c1 = df1['gender'].value_counts()
print(c1)
```

```python
# Pie Chart to show Gender Distribution Between Candidates
plt.pie(c1,labels = ['Male','Female'], autopct = "%0.2f%%", explode=[0,0.2], shadow = True)
plt.legend()
```

```python
# Boxplot
gen=df1.groupby('gender').apply(lambda x:x['name'].count()).reset_index(name='counts')
winners=df1[df1['winner']==1]
gen_win=winners.groupby('gender').apply(lambda x:x['name'].count()).reset_index(name='counts')
total=pd.concat([gen_win,gen])
sns.barplot(x='gender',y='counts',data = total, palette="YlGnBu")
```

```python
state_criminal = df1.groupby('state')[['criminal_cases']].sum().sort_values(by=['criminal_cases']).sort_values(by=['state'])
print(state_criminal)
x = state_criminal['criminal_cases'].tolist()
print(x)
```

```python
# Winning Candidates in each State with Criminal Cases
state_criminal_winner = df1[df1['winner']>0].groupby('state')[['criminal_cases']].sum().sort_values(by=['criminal_cases']).sort_values(by=['state'])
print(state_criminal_winner)
y = state_criminal_winner['criminal_cases'].tolist()
print(y)
```

```python
# Barplot to Represent above Data
state = df['state'].unique().tolist()
xpos = np.arange(len(state_criminal))
plt.xticks(xpos,state)
plt.xticks(rotation =90)
plt.bar(xpos,x)
```

```python
plt.xticks(xpos,state)
plt.xticks(rotation =90)
plt.bar(xpos,y)
```

```python
# Dual Bar graph to represent Criminal Case and Winner in each state 
plt.xticks(xpos,state)
plt.bar(xpos-0.2,x,width=0.4, label = 'Total Cases')
plt.bar(xpos+0.2,y,width=0.4, label = 'Winner')

plt.ylabel('No. of Cases')
plt.xticks(rotation = 90)
plt.legend()
```

```python
# Scatter Plot on State vs Constituency
party_state=df1.groupby('party')['state'].nunique() .reset_index(name='state')
party_const=df1.groupby('party')['constituency'].count().reset_index(name='constituency')
top_const=party_const.sort_values(by='constituency',ascending=False)
top_party=pd.merge(top_const,party_state,how='inner',on='party')
sns.scatterplot(x = 'constituency', y = 'state', data = top_party, hue='state', size='constituency')
```

```python
st_party=winners.groupby(['party','state'])['winner'].sum().reset_index(name='wins')
print(st_party)
pivot_st_party=pd.pivot(st_party,index='party',columns='state',values='wins')
plt.figure(figsize=(15,35))
sns.heatmap(pivot_st_party,annot=True, cmap="YlGnBu")
plt.xlabel('state')
plt.ylabel('party')
plt.title('Statewise report card for the Political Parties in India',size=25)
```

```python
# Category Wise Candidates
df1 = df[df['party']!='NOTA']
d = df1.groupby(['category']).count()
list1 =d.index.tolist()
list2 = d['name'].tolist()
sns.barplot(x = list1, y= list2)
```

```python
# Relation in  total candidates and winner on basis of Category
df1 = df1[df1['winner']==1]
d1 = df1.groupby(['category']).count()
ind = d1.index.tolist()
list1=d1['name'].tolist()
xpos = np.arange(len(list1))
list2 = d['name'].tolist()
plt.xticks(xpos,ind)
plt.bar(xpos-0.2,list1,width=0.4, label = 'Overall Category')
plt.bar(xpos+0.2,list2,width=0.4, label = 'Winner')
plt.legend()
```
